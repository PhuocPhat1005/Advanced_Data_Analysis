```python
import os
import pandas as pd
import numpy as np
from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
def setup_gemini_agent(df, temperature=0, model="gemini-1.5-flash"):
    llm = ChatGoogleGenerativeAI(
        model=model,
        temperature=temperature,
    )

    # 2) Khởi agent, đẩy các kwargs vào agent_executor_kwargs
    agent = create_pandas_dataframe_agent(
        llm=llm,
        df=df,
        verbose=True,
        allow_dangerous_code=True,
        return_intermediate_steps=True,
        agent_executor_kwargs={
            "handle_parsing_errors": True
        },
    )
    return agent
```

```python
agent = setup_gemini_agent(df)
```

```python
instruction = f"Hãy cho tôi biết tên sản phẩm có giá cao nhất trong Products.csv"
```

```python
results = agent.invoke(instruction)
```
